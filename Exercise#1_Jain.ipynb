```python
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "C:/Users/Chira/OneDrive/Documents/Winter 2025/AIDI 2000 - Applied Machine Learning/shopping_trends.csv"
df = pd.read_csv(file_path)
```

```python
#Let's explore the data
df.head()
```

```python
df.info() #get info about the columns
```

```python
#Check for missing values
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()
```

```python

low_end = df['Age'].min()
high_end = df['Age'].max()

print(f"Low end (minimum value): {low_end}")
print(f"High end (maximum value): {high_end}")
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)



print(f"Q1 (25th Percentile): {Q1}")
print(f"Q3 (75th Percentile): {Q3}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Purchase Amount (USD)'])
plt.title('Boxplot of Purchase Amount')
plt.show()
```

```python
location_counts = df['Location'].value_counts().head(10)
sns.barplot(x=location_counts.values, y=location_counts.index, palette='coolwarm')
plt.title('Top 10 Locations')
plt.show()
```

```python
payment_methods = df['Payment Method'].value_counts()
sns.barplot (x= payment_methods.values, y = payment_methods.index, palette = 'coolwarm')
plt.title ("Top Payment Methods")

plt.show()
```

```python
low_end = df['Purchase Amount (USD)'].min()
high_end = df['Purchase Amount (USD)'].max()

print(f"Low end (minimum value): {low_end}")
print(f"High end (maximum value): {high_end}")
```

```python
print(f"Number of unique Customer IDs: {df['Customer ID'].nunique()}")
duplicates = df['Customer ID'].duplicated().sum()
print(f"Number of duplicate Customer IDs: {duplicates}")
```

```python
gender_counts = df['Gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='Set2')
plt.title('Gender Comparison')
plt.ylabel('Count')
plt.show()
```

```python
top_categories = df['Category'].value_counts()
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Categories by largest to smallest')
plt.xlabel('Count')
plt.show()
```

```python
print(df['Purchase Amount (USD)'].describe())
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])


correlation_matrix = numeric_df.corr()
print(correlation_matrix)


```

```python


plt.figure(figsize=(12, 8))


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)


plt.title("Correlation Matrix", fontsize=16)
plt.show()
```
